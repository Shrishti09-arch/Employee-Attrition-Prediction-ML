# 23. Data Preprocessing Pipeline

Machine learning algorithms require numerical input data.

To prepare the dataset, two preprocessing techniques are applied:

- **StandardScaler** standardizes numerical features so that they have a mean of zero and a standard deviation of one.
- **OneHotEncoder** converts categorical variables into numerical binary features.

Instead of performing these transformations manually, a **ColumnTransformer** is used to automate the preprocessing process.

This approach improves reproducibility, prevents data leakage, and follows industry best practices.

```python
# ==========================================
# Preprocessing Pipeline
# ==========================================

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler

preprocessor = ColumnTransformer(

    transformers=[

        ('num', StandardScaler(), numerical_features),

        ('cat', OneHotEncoder(drop='first', handle_unknown='ignore'), categorical_features)

    ]

)

print("Preprocessing pipeline created successfully.")
```

### Observation

The preprocessing pipeline has been successfully created.

Numerical features will be standardized to improve model performance, while categorical variables will be converted into binary indicator variables using One-Hot Encoding.

Using a `ColumnTransformer` ensures that each preprocessing technique is applied only to the appropriate type of feature.

# 24. Apply Feature Transformation

The preprocessing pipeline is now fitted using the training data.

The learned transformations are then applied to both the training and testing datasets.

Fitting the preprocessing pipeline only on the training data prevents information leakage from the test set and ensures a fair evaluation of model performance.

```python
# ==========================================
# Apply Preprocessing
# ==========================================

X_train_processed = preprocessor.fit_transform(X_train)

X_test_processed = preprocessor.transform(X_test)

print("Training Data Shape :", X_train_processed.shape)
print("Testing Data Shape  :", X_test_processed.shape)
```

### Observation

The preprocessing pipeline has successfully transformed both the training and testing datasets.

The increase in the number of features is expected because categorical variables have been expanded into multiple binary variables through One-Hot Encoding.

The processed datasets now contain only numerical values and are ready for machine learning model development.

# 27. Attrition Rate by Department

Departments within an organization often experience different levels of employee turnover.

This analysis calculates the percentage of employees who leave from each department.

Understanding department-wise attrition enables HR teams to prioritize retention efforts where they are needed most.

```python
# ==========================================
# Attrition Rate by Department
# ==========================================

department_attrition = (
    df.groupby("Department")["Attrition"]
      .agg(
          Total_Employees="count",
          Employees_Left="sum",
          Attrition_Rate="mean"
      )
)

department_attrition["Attrition_Rate"] = (
    department_attrition["Attrition_Rate"] * 100
).round(2)

department_attrition = department_attrition.sort_values(
    by="Attrition_Rate",
    ascending=False
)

department_attrition
```

```python
plt.figure(figsize=(8,5))

department_attrition.plot(
    kind='bar',
    edgecolor='black'
)

plt.title("Attrition Rate by Department")
plt.xlabel("Department")
plt.ylabel("Attrition Rate (%)")
plt.xticks(rotation=0)

plt.tight_layout()

plt.savefig("charts/chart1_department_attrition.png", dpi=300)

plt.show()
```

### Observation

The analysis reveals that employee attrition varies across departments.

Departments with higher attrition rates should receive greater attention from HR through improved engagement programs, workload analysis, career development opportunities, and employee satisfaction initiatives.

This visualization helps identify departments requiring immediate retention strategies.

#  Machine Learning Model Building

After understanding the dataset through Exploratory Data Analysis (EDA), the next step is to build predictive machine learning models.

The objective is to predict whether an employee is likely to leave the organization based on employee-related attributes.

Three classification algorithms will be developed and compared:

- Logistic Regression
- Random Forest Classifier
- Gradient Boosting Classifier

To ensure fair evaluation, the dataset will be divided into training and testing sets. Since employee attrition is an imbalanced classification problem, class weights will be used where applicable to improve prediction of minority class employees.

## Preparing Features and Target Variable

Machine learning models require separating the dataset into:

- Features (X): Independent variables used for prediction.
- Target (y): Employee Attrition (0 = Stayed, 1 = Left).

This separation ensures that the model learns patterns from employee characteristics while predicting the target variable.

```python
# ====================================
# Preparing Features and Target
# ====================================

X = df.drop("Attrition", axis=1)

y = df["Attrition"]

print("Shape of Features :", X.shape)
print("Shape of Target   :", y.shape)
```

### Observation

The independent variables have been separated from the target variable.

The machine learning algorithms will use the features to learn patterns associated with employee attrition.

## Encoding Categorical Variables

Most machine learning algorithms cannot directly process categorical data.

Therefore, One-Hot Encoding is applied to convert categorical columns into numerical binary variables while preserving all available information.

```python
# ====================================
# One-Hot Encoding
# ====================================

X = pd.get_dummies(
    X,
    drop_first=True
)

print("Shape after Encoding:", X.shape)
```

### Observation

Categorical variables have been successfully transformed into numerical features.

This enables machine learning models to process all employee information efficiently.

## Splitting Training and Testing Data

The dataset is divided into training and testing subsets.

- Training Set: 80%
- Testing Set: 20%

The training data is used to build the models, while the testing data evaluates their ability to generalize to unseen employees.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Ensure all column names are strings
X.columns = X.columns.astype(str)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

# Ensure train/test column names are strings
X_train.columns = X_train.columns.astype(str)
X_test.columns = X_test.columns.astype(str)

# Scale the features
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### Observation

Using stratified sampling ensures that both training and testing datasets preserve the original proportion of employees who stayed and employees who left.

This results in a fair and unbiased evaluation.

## Feature Scaling

Many machine learning algorithms perform better when numerical features are standardized.

StandardScaler transforms the features to have:

- Mean = 0
- Standard Deviation = 1

Scaling improves convergence for algorithms such as Logistic Regression while maintaining consistency across features.

```python
from sklearn.preprocessing import StandardScaler

# ====================================
# Standard Scaling
# ====================================

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)

print("Scaling Completed Successfully")
```

### Observation

Training data has been standardized and the same transformation has been applied to the testing data.

This prevents data leakage and ensures fair model evaluation.

## Importing Classification Algorithms

Three supervised learning algorithms will be trained and compared:

1. Logistic Regression
2. Random Forest
3. Gradient Boosting

Each algorithm has different strengths, allowing us to identify the most suitable model for employee attrition prediction.

```python
from sklearn.linear_model import LogisticRegression

from sklearn.ensemble import (
    RandomForestClassifier,
    GradientBoostingClassifier
)

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score
)
```

## Model 1: Logistic Regression

Logistic Regression is a supervised machine learning algorithm commonly used for binary classification problems.

It serves as a strong baseline model due to its simplicity, interpretability, and ability to estimate the probability of employee attrition.

Since the dataset is slightly imbalanced, the `class_weight='balanced'` parameter is used to improve prediction performance for the minority class.

```python
# ==============================
# Import Logistic Regression
# ==============================

from sklearn.linear_model import LogisticRegression

# ==============================
# Initialize the Model
# ==============================

lr_model = LogisticRegression(
    class_weight='balanced',
    random_state=42,
    max_iter=1000
)

# ==============================
# Train the Model
# ==============================

lr_model.fit(X_train_scaled, y_train)

# ==============================
# Make Predictions
# ==============================

y_test_lr_pred = lr_model.predict(X_test_scaled)

# ==============================
# Prediction Probabilities
# ==============================

y_test_lr_prob = lr_model.predict_proba(X_test_scaled)[:,1]

print("Logistic Regression model trained successfully!")
```

### Observation

The Logistic Regression model was successfully trained using the scaled training dataset.

It serves as the baseline model for comparison against more advanced ensemble learning algorithms.

## Model 2: Random Forest Classifier

Random Forest is an ensemble learning algorithm that combines multiple decision trees to improve predictive accuracy and reduce overfitting.

Unlike Logistic Regression, Random Forest can capture complex non-linear relationships between employee characteristics and attrition.

The `class_weight='balanced'` parameter is used to compensate for class imbalance within the dataset.

```python
# ==============================
# Import Random Forest
# ==============================

from sklearn.ensemble import RandomForestClassifier

# ==============================
# Initialize the Model
# ==============================

rf_model = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    class_weight='balanced'
)

# ==============================
# Train the Model
# ==============================

rf_model.fit(X_train, y_train)

# ==============================
# Make Predictions
# ==============================

y_test_rf_pred = rf_model.predict(X_test)

# ==============================
# Prediction Probabilities
# ==============================

y_test_rf_prob = rf_model.predict_proba(X_test)[:,1]

print("Random Forest model trained successfully!")
```

### Observation

The Random Forest model was successfully trained using the original training dataset.

This ensemble model is expected to outperform Logistic Regression by capturing complex interactions among employee-related features.

## Model 3: Gradient Boosting Classifier

Gradient Boosting builds trees sequentially by correcting previous prediction errors.

It often achieves higher predictive performance by focusing on difficult-to-classify observations.

```python
# ==============================
# Import Gradient Boosting
# ==============================

from sklearn.ensemble import GradientBoostingClassifier

# ==============================
# Initialize the Model
# ==============================

gb_model = GradientBoostingClassifier(
    random_state=42
)

# ==============================
# Train the Model
# ==============================

gb_model.fit(X_train, y_train)

# ==============================
# Make Predictions
# ==============================

y_test_gb_pred = gb_model.predict(X_test)

# ==============================
# Prediction Probabilities
# ==============================

y_test_gb_prob = gb_model.predict_proba(X_test)[:,1]

print("Gradient Boosting model trained successfully!")
```

### Observation

All three machine learning models have been successfully trained and are ready for performance evaluation.

## Comparing Model Performance

Instead of evaluating models individually, a comparison table provides a concise summary of model performance across multiple evaluation metrics.

This makes it easier to identify the best-performing algorithm.

```python
results = pd.DataFrame({

    "Model":[
        "Logistic Regression",
        "Random Forest",
        "Gradient Boosting"
    ],

    "Accuracy":[
        accuracy_score(y_test,lr_pred),
        accuracy_score(y_test,rf_pred),
        accuracy_score(y_test,gb_pred)
    ],

    "Precision":[
        precision_score(y_test,lr_pred),
        precision_score(y_test,rf_pred),
        precision_score(y_test,gb_pred)
    ],

    "Recall":[
        recall_score(y_test,lr_pred),
        recall_score(y_test,rf_pred),
        recall_score(y_test,gb_pred)
    ],

    "F1 Score":[
        f1_score(y_test,lr_pred),
        f1_score(y_test,rf_pred),
        f1_score(y_test,gb_pred)
    ],

    "ROC-AUC":[
        roc_auc_score(y_test,lr_prob),
        roc_auc_score(y_test,rf_prob),
        roc_auc_score(y_test,gb_prob)
    ]

})

results = results.round(3)

results.sort_values(
    by="ROC-AUC",
    ascending=False
)
```

### Observation

The comparison table enables objective evaluation of all three models using multiple performance metrics rather than relying solely on accuracy.

The best-performing model will be selected based on a combination of Precision, Recall, F1-Score, and ROC-AUC.

## Detailed Classification Reports

Classification reports provide class-wise evaluation including Precision, Recall and F1-Score.

These metrics are particularly useful for understanding how well each model predicts employee attrition.

```python
from sklearn.metrics import classification_report

models = {
    "Logistic Regression": lr,
    "Random Forest": rf,
    "Gradient Boosting": gb
}

for name, model in models.items():

    predictions = model.predict(X_test)

    print("="*70)
    print(name)
    print("="*70)

    print(classification_report(
        y_test,
        predictions
    ))
```

### Observation

The classification reports reveal how accurately each model identifies employees who stayed and those who left.

Special attention should be given to Recall for the Attrition class because missing employees likely to leave may reduce the usefulness of the prediction system.

## Confusion Matrix

A confusion matrix provides a visual summary of prediction performance.

It shows:

- True Positives
- True Negatives
- False Positives
- False Negatives

An ideal model maximizes correct predictions while minimizing errors.

```python
from sklearn.metrics import ConfusionMatrixDisplay

fig, axes = plt.subplots(1,3,figsize=(18,5))

for ax, (name, model) in zip(axes, models.items()):

    ConfusionMatrixDisplay.from_estimator(
        model,
        X_test,
        y_test,
        cmap="Blues",
        ax=ax,
        colorbar=False
    )

    ax.set_title(name)

plt.tight_layout()

plt.savefig("charts/confusion_matrices.png", dpi=300)

plt.show()
```

### Observation

The confusion matrices illustrate the number of correctly classified employees and prediction errors for each model.

A strong model should correctly identify both employees who remain and employees likely to leave while minimizing false predictions.

## ROC Curve Comparison

Receiver Operating Characteristic (ROC) curves compare the classification capability of all models across different decision thresholds.

The closer the curve is to the upper-left corner, the better the classifier performs.

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import auc

plt.figure(figsize=(8,6))

for name, model in models.items():

    probabilities = model.predict_proba(X_test)[:,1]

    fpr, tpr, _ = roc_curve(
        y_test,
        probabilities
    )

    roc_auc = auc(fpr, tpr)

    plt.plot(
        fpr,
        tpr,
        linewidth=2,
        label=f"{name} (AUC = {roc_auc:.3f})"
    )

plt.plot(
    [0,1],
    [0,1],
    linestyle="--"
)

plt.xlabel("False Positive Rate")

plt.ylabel("True Positive Rate")

plt.title("ROC Curve Comparison")

plt.legend()

plt.grid(alpha=0.3)

plt.tight_layout()

plt.savefig("charts/roc_curve.png", dpi=300)

plt.show()
```

### Observation

The ROC curves demonstrate the discriminative ability of each model.

A larger Area Under the Curve (AUC) indicates better capability to distinguish employees likely to leave from those expected to remain.
